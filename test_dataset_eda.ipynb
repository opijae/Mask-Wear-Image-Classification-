```python
csv_list=os.listdir('./outputs')
temp=pd.DataFrame()
for i,csv_file in enumerate(csv_list):
    # csv_data=np.loadtxt(f'file_all_predictions_{i}.csv',delimiter=',',dtype=np.float64)
    # csv_data_1=np.loadtxt(f'file_all_predictions_efficientnet_0.001_{i}.csv',delimiter=',',dtype=np.float64)
    # csv_data_2=(csv_data_1+csv_data)/2
    if csv_file=='output_real_2.csv':
        continue
    csv_path=os.path.join('./outputs',csv_file)
    print(csv_path)
    # csv_data=np.loadtxt(csv_path,delimiter=',',dtype=np.float64)
    info = pd.read_csv(csv_path)
    try:
        temp[csv_file[4:]]=info['ans']
    except:
        try:
            temp[csv_file]=info['0']
        except:
            continue
```
